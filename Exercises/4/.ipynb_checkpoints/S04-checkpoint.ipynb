```python
# Bias-Variance Decomposition. No noise (sig=0)

import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt

m = 10     # size of training set
L = 100    # number of training sets
p = 1000   # size of test set

sig = 0.   # standard deviation of Gaussian noise

x_test = np.random.rand(p)*2-1
y_test = np.tanh(10*x_test) + sig*np.random.randn(p)


REG = [10, 1, 0.5, 0.1, 0.05, 0.02, 0.01, 0.009, 0.005, 0.001]

X_train = np.random.rand(L,m)*2-1
Y_train = np.tanh(10*X_train) + sig*np.random.randn(L,m)

H = np.zeros((L,p)) # for storing all hypothesis evaluations at test set

# the following lists contain the quantities for each regularization parameter
BIAS = []
VAR = []
TEST_ERROR = []

for reg in REG:
    model = Pipeline([('poly', PolynomialFeatures(degree=9)),('linear', Ridge(alpha=reg, fit_intercept=True))])
    for l in range(L):
        x = X_train[l,:]
        y = Y_train[l,:]
        model = model.fit(x[:, np.newaxis], y)
        
        h = model.predict(x_test.reshape((p,1))) # we need h_star on testset, so evaluate each h on test set
        H[l,:] = h

    h_star = np.mean(H,axis=0) # mean predictions on test set
    
    truth = np.tanh(10*x_test) # evaluate truth on test set
    
    # bias
    bias = np.mean((truth - h_star)**2)
    BIAS.append(bias)
    
    # variance
    var = 0
    for l in range(L):
        var += np.mean((H[l,:] - h_star)**2) / L
    VAR.append(var)
    
    # expected test error. performance averaged over multiple training datasets
    e_test = 0
    for l in range(L):
        e_test += np.mean((H[l,:] - y_test)**2) / L
    TEST_ERROR.append(e_test)
    
    
ind = np.argmin(TEST_ERROR)
print('Best regularization parameter is lambda = ', REG[ind])
    
plt.plot(REG,BIAS, label='bias')
plt.plot(REG,VAR, label='variance')
plt.plot(REG,np.add(BIAS,VAR), label='bias+variance')
plt.plot(REG,TEST_ERROR, '--', label='test error')
plt.plot( [REG[ind],REG[ind]], [0.0001,0.6], 'k--', label='best $\lambda$' )
plt.legend()
plt.xlabel('regularization parameter $\lambda$')
plt.xscale('log')
plt.show()

print('TEST - (BIAS+VAR) = ',TEST_ERROR - (np.array(BIAS) + VAR))
print('MEAN(TEST - (BIAS+VAR)) = ', np.mean(TEST_ERROR - (np.array(BIAS) + VAR)))
print('TEST - (BIAS+VAR) = ',np.array(TEST_ERROR[ind] - (BIAS[ind] + VAR[ind])) )
```

```python
# Bias-Variance Decomposition. With noise (sig=0.5)
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt

m = 10     # size of training set
L = 100    # number of training sets
p = 1000   # size of test set

sig = 0.5

x_test = np.random.rand(p)*2-1
y_test = np.tanh(10*x_test) + sig*np.random.randn(p)


REG = [10, 1, 0.5, 0.1, 0.05, 0.02, 0.01, 0.009, 0.005]

X_train = np.random.rand(L,m)*2-1
Y_train = np.tanh(10*X_train) + sig*np.random.randn(L,m)

H = np.zeros((L,p)) # all hypotheses evaluated at test set

BIAS = []
VAR = []
TEST_ERROR = []

for reg in REG:
    model = Pipeline([('poly', PolynomialFeatures(degree=9)),('linear', Ridge(alpha=reg, fit_intercept=True))])
    for l in range(L):
        x = X_train[l,:]
        y = Y_train[l,:]
        model = model.fit(x[:, np.newaxis], y)
        
        h = model.predict(x_test.reshape((p,1))) # we need h_star on testset, so evaluate each h on test set
        H[l,:] = h

    h_star = np.mean(H,axis=0)
    
    truth = np.tanh(10*x_test) # evaluate truth on test set
    
    # bias
    bias = np.mean((truth - h_star)**2)
    BIAS.append(bias)
    
    # variance
    var = 0
    for l in range(L):
        var += np.mean((H[l,:] - h_star)**2) / L
    VAR.append(var)
    
    # expected test error. performance averaged over multiple training datasets
    e_test = 0
    for l in range(L):
        e_test += np.mean((H[l,:] - y_test)**2) / L
    TEST_ERROR.append(e_test)
    
    #print(bias)
    #print(var)
    #print(e_test)
    
ind = np.argmin(TEST_ERROR)
print('Best regularization parameter is lambda = ', REG[ind])
    
plt.plot(REG,BIAS, label='bias')
plt.plot(REG,VAR, label='variance')
plt.plot(REG,np.add(BIAS,VAR), label='bias+variance')
plt.plot(REG,TEST_ERROR, '--', label='test error')
plt.plot( [REG[ind],REG[ind]], [0.0001,1.5], 'k--', label='best $\lambda$' )
plt.legend()
plt.xlabel('regularization parameter $\lambda$')
plt.xscale('log')
plt.show()

print('All deviations:')
print('TEST - (BIAS+VAR) = ',TEST_ERROR - (np.array(BIAS) + VAR))
print('')
print('Deviation for best regularization parameter:')
print('TEST - (BIAS+VAR) = ',np.array(TEST_ERROR[ind] - (BIAS[ind] + VAR[ind])) )
```

Discussion 
* Caution: Model complexity increases to the left (with decreasing regularization parameter)
* The deviation is zero for sig=0 and 0.25 = sig^2 for sig=0.5. This is due to: Err = bias + var + sig^2
