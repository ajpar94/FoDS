# Regression functions

```python
# These functions can be used to fit the models.
# You can change anything you want and you can use and add your own functions.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

def poly_basis_transform(X,k):
    """
    X: one dimensional input array
    k: order of polynomial
    
    returns
    PHI: data matrix with polynomial features
    """
    X = np.array(X)
    m = X.size
    PHI = np.ones((m,k+1))
    for i in range(k+1):
        PHI[:,i] = np.power(X,i)
    return PHI


def mse(PHI,y,w):
    """
    PHI: augmented feature matrix
    y:   output vector
    w:   weight vector
    
    returns mean squared error
    """
    m = PHI.shape[0]
    return 1/(2*m) * np.linalg.norm(PHI.dot(w) - y)**2


def linear_eval(PHI,w):
    """
    PHI: augmented feature matrix
    w:   weight vector
    
    returns: array of linear hypothesis evaluations h(x) for all examples x in PHI
    """
    return PHI.dot(w)


def linear_regression_fit(PHI,y,n_iter,eta,lamb=0,w_init=None):
    """
    PHI:    augmented feature matrix
    y:      output vector
    n_iter: number of gradient descent iterations
    eta:    learning rate
    lamb:   regularization parameter
    w_init: initial weights
    
    returns
    w:    fitted weight vector
    loss: list of MSE values for each iteration
    """
    (m,k) = PHI.shape
    
    if w_init is None:
        w_init = 0.1*np.random.randn(k)
    
    w = w_init
    loss = [mse(PHI,y,w)] # first entry of loss is error with respect to initial weights
    
    for i in range(n_iter):
        w0 = w.copy()
        w0[0] = 0     # exclude bias from regularization
        grad = 1/m * (PHI.T.dot(PHI.dot(w) - y) + lamb*w0)
        #if (np.linalg.norm(eta * grad) < 1e-7):
            #break
        w = w - eta*grad
        loss.append(mse(PHI,y,w))
        
    return w,loss


def leading_ones(X):
    n,_ = X.shape
    return np.hstack(((np.ones(n))[:, np.newaxis], X))
```

# Exercise 3.1

```python
test_data = np.genfromtxt('f31.csv', delimiter=";")
Y_test = test_data[:,2]
PHI_test = leading_ones(test_data[:,:2])



def experiment(m):
    """
    Generate training set with m samples
    Perform multiple linear regression
    Compute training error of fitted model
    Estimate generalization error of fitted model on test_data
    """
    global Y_test, PHI_test
    
    # create training data
    X = (np.random.rand(m,2)*2)-1
    PHI = leading_ones(X)
    # compute output labels (x1*sin(pi*x2)) 
    Y = X[:,0]* np.sin(np.pi*X[:,1])
    #print(np.hstack((X,Y[:,np.newaxis])))
    
    # fit model through linear regression
    w, loss = linear_regression_fit(PHI,Y,100000, 1e-4)
    
    #training error
    training_error = loss[-1]
    #generalization error
    gen_error = mse(PHI_test,Y_test,w)
    
    return (training_error, gen_error)


def ex31():
    m_list = range(2,81)
    # final list for plotting
    training_errors = []
    generalization_errors = []

    for m in m_list:
        print(m)
        tr_error_temp = 0
        g_error_temp = 0
        # calculate average over 100 trials
        for _ in range(100):
            # perform experiment
            tr_error, g_error = experiment(m)
            
            tr_error_temp += tr_error
            g_error_temp += g_error
            
        training_errors.append(tr_error_temp/100)
        generalization_errors.append(g_error_temp/100)

    #plotting training_errors, generalization_errors
    plt.figure(figsize=(20, 10))
    plt.suptitle("Comparison of training and generalization errors",
                  size=20, fontweight='bold')
    plt.xlabel("sample size m", size=20)
    plt.ylabel("MSE", size=20)
    plt.plot(training_errors, label="Training errors")
    plt.plot(generalization_errors, label="Generalization error")
    plt.legend()
    plt.show()
    
    return training_errors, generalization_errors
```

```python
ex31()
```

# Exercise 3.2

```python
train_data2 = np.genfromtxt('f32train.csv', delimiter=";")[:,:2]
Y_train2 = train_data2[:,1]
test_data2 = np.genfromtxt('f32test.csv', delimiter=";")[:,:2]
Y_test2 = test_data2[:,1]

def poly_reg(k):
    global train_data2, Y_train2, test_data2, Y_test2
    
    PHI_train2 = poly_basis_transform(train_data2[:,0],k)
    # multiple linear regression
    w, loss = linear_regression_fit(PHI_train2,Y_train2,100000, 1e-2)
    # generalization error
    PHI_test2 = poly_basis_transform(test_data2[:,0],k)
    gen_error = mse(PHI_test2,Y_test2,w)
    
    return gen_error


def crossval(k,folds=10):
    global train_data2
    
    splits = np.vsplit(train_data2,folds)
    
    gen_error = 0
    for f in range(folds):
        splits_temp = splits.copy()
        test = splits_temp.pop(f)
        train = np.vstack(splits_temp)
        
        Xtest, Ytest = test[:,0], test[:,1]
        Xtrain, Ytrain = train[:,0], train[:,1]
        
        PHItrain = poly_basis_transform(Xtrain,k)
        # multiple linear regression
        w, loss = linear_regression_fit(PHItrain,Ytrain,100000, 1e-2)
        # generalization error
        PHItest = poly_basis_transform(Xtest,k)
        gen_error += mse(PHItest,Ytest,w)
        
    return gen_error/folds


def ex32():
    
    p_list = []
    c_list = []
    
    for  k in range(10):
        k+=1
        print(k)
        p_list.append(poly_reg(k))
        c_list.append(crossval(k))
    
    #plotting
    plt.figure(figsize=(20, 10))
    plt.suptitle("Comparison of test-MSE and cross validation errors",
                  size=20, fontweight='bold')
    plt.xlabel("order k", size=20)
    plt.ylabel("MSE", size=20)
    plt.plot(p_list, label="test-MSE")
    plt.plot(c_list, label="cross validation error")
    plt.legend(fontsize=20)
    plt.show()
    
    return p_list, c_list
```

# Exercise 3.3

```python
train_data3 = np.genfromtxt('f33.csv', delimiter=";")[:,:2]
Y_train3 = train_data3[:,1]

def nested_crossval(k=10,outer_folds=5,inner_folds=4):
    global train_data3
    
    #print(train_data3.shape)
    outer_splits = np.vsplit(train_data3,outer_folds)
    gen_oerror = 0
    for of in range(outer_folds):
        print(of)
        outer_splits_temp = outer_splits.copy()
        otest = outer_splits_temp.pop(of)
        otrain = np.vstack(outer_splits_temp)
        #print(otrain.shape)
        
    
        splits = np.vsplit(otrain,inner_folds)

        weights = []
        errors = []
        for lam in [0.01, 0.03, 0.1, 0.3, 1.0]:
        
            gen_error = 0
            for f in range(inner_folds):
                splits_temp = splits.copy()
                test = splits_temp.pop(f)
                train = np.vstack(splits_temp)

                Xtest, Ytest = test[:,0], test[:,1]
                Xtrain, Ytrain = train[:,0], train[:,1]

                PHItrain = poly_basis_transform(Xtrain,k)
                # multiple linear regression
                w, loss = linear_regression_fit(PHItrain,Ytrain,100000, 1e-2, lam)
                # generalization error
                PHItest = poly_basis_transform(Xtest,k)
                gen_error += mse(PHItest,Ytest,w)

            weights.append(w)
            errors.append(gen_error/inner_folds)
        
        w_best = weights[np.argmin(errors)]
        
        Xotest, Yotest = otest[:,0], otest[:,1]
        PHIotest = poly_basis_transform(Xotest,k)
        gen_oerror += mse(PHIotest,Yotest,w_best)
        
    return gen_oerror/outerfolds
```

```python
nested_crossval()
```
